```python
import os
import sys
sys.path.append("/kaggle/input/automated-essay-scoring")

import pandas as pd
import numpy as np

from linguistic_features import get_features

data_dir = '/kaggle/input/learning-agency-lab-automated-essay-scoring-2/train.csv'
data_extra_dir = "/kaggle/input/persaude-corpus-2/persuade_2.0_human_scores_demo_id_github.csv"
output_dir = "/kaggle/working/"
```

```python
train_df = pd.read_csv(data_dir)

get_features(train_df, True, os.path.join('/kaggle/working', 'train_linguistic.csv'))
```

```python
extra_data_df = pd.read_csv(data_extra_dir)

extra_data_df = extra_data_df.rename(columns={'essay_id_comp': 'essay_id',
                                              'holistic_essay_score': 'score'})

extra_data_df = extra_data_df[['essay_id', 'full_text', 'score']]
extra_data_df = extra_data_df[~extra_data_df['full_text'].isin(train_df['full_text'])]

get_features(extra_data_df, True, os.path.join('/kaggle/working', 'extra_linguistic.csv'))
```

```python
import shutil

dont_removes = ['train_linguistic.csv', 'extra_linguistic.csv', 'features.txt']

for f in os.listdir('/kaggle/working'):
    if f not in dont_removes:
        if os.path.isfile(f):
            os.remove(f)
        else:
            shutil.rmtree(f)
```
